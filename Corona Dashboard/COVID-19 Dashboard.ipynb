```python
import bokeh
import pandas as pd
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
from bokeh.models import DatetimeTickFormatter,HoverTool
output_notebook()
```

```python
dly_data = pd.read_csv('covid_19_india.csv')
```

```python
dly_data.head()
```

```python
dly_data.rename(columns={'State/UnionTerritory':'State'},inplace=True)
```

```python
dly_data['Date'] = dly_data['Date'].map(lambda x:x.replace('/2020','/20'))
```

```python
dly_data['Date'] = pd.to_datetime(dly_data['Date'],format='%d/%m/%y').dt.strftime('%d-%b-%y')
```

```python
dly_data['Month'] = dly_data['Date'].map(lambda x:x.split('-')[1])
```

```python
Mnthy_confmed_cse= dly_data.groupby(['Date','State']).agg({'Confirmed':'max'}).reset_index() 
```

```python
Overall_trend = Mnthy_confmed_cse.groupby(['Date']).agg({'Confirmed':'sum'}).reset_index() 
Overall_trend.sort_values(by='Confirmed',ascending = True,inplace=True)
Overall_trend['Date'] = Overall_trend['Date'].astype('datetime64[ns]') 
```

```python
x=list(Overall_trend['Date'])
y=list(Overall_trend['Confirmed'])
p = figure(title="",plot_width=1000, plot_height=500,x_minor_ticks='auto',x_axis_label='Date'
           ,y_axis_label='Number of people Affected')
```

```python
p.line(x,y,line_width = 3)
hover = p.select(dict(type=HoverTool))
hover.tooltips = [
        ("Date", "$x"),
        ("Value", "$y"),
        ]
```

```python
p.xaxis.formatter=DatetimeTickFormatter(
        hours=["%d-%b-%y"],
        days=["%d-%b-%y"],
        months=["%d-%b-%y"],
        years=["%d-%b-%y"]
)
show(p)
```

```python
from jupyter_plotly_dash import JupyterDash

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

app = JupyterDash('SimpleExample')

app.layout = html.Div([
    dcc.RadioItems(
        id='dropdown-color',
        options=[{'label': c, 'value': c.lower()}
                 for c in ['Red', 'Green', 'Blue']],
        value='red'
    ),
    html.Div(id='output-color'),
    dcc.RadioItems(
        id='dropdown-size',
        options=[{'label': i, 'value': j}
                 for i, j in [('L','large'), ('M','medium'), ('S','small')]],
        value='medium'
    ),
    html.Div(id='output-size')

])

@app.callback(
    dash.dependencies.Output('output-color', 'children'),
    [dash.dependencies.Input('dropdown-color', 'value')])
def callback_color(dropdown_value):
    return "The selected color is %s." % dropdown_value

@app.callback(
    dash.dependencies.Output('output-size', 'children'),
    [dash.dependencies.Input('dropdown-color', 'value'),
     dash.dependencies.Input('dropdown-size', 'value')])
def callback_size(dropdown_color, dropdown_size):
    return "The chosen T-shirt is a %s %s one." %(dropdown_size,
                                                  dropdown_color)

app
```
