```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
import time
import os
from selenium.webdriver.common.keys import Keys
```

```python
def findelement():
    print('findelement')
    time.sleep(2)
    role = driver.find_element_by_xpath('//div[@class="list"]').find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
    #lrole.append(role.text)
    time.sleep(2)
    company = driver.find_element_by_xpath('//div[@class="list"]').find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
    #lcompany.append(company.text)
    time.sleep(2)
    exp = driver.find_element_by_xpath('//div[@class="list"]').find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')
    #lexp.append(exp.text)
    time.sleep(2)
    salary = driver.find_element_by_xpath('//div[@class="list"]').find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi salary"]')
    #lsalary.append(salary.text)
    time.sleep(2)
    location = driver.find_element_by_xpath('//div[@class="list"]').find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
    #llocation.append(location.text)
    time.sleep(2)
    desc = driver.find_element_by_xpath('//div[@class="list"]').find_elements_by_xpath('//div[@class="job-description fs12 grey-text"]')
    #ldesc.append(desc.text)
    time.sleep(2)
    req = driver.find_element_by_xpath('//div[@class="list"]').find_elements_by_xpath('//ul[@class="tags has-description"]')
    #lreq.append(req.text)
    
    for i in role:
        lrole.append(i.text)
            
    for j in company:
        lcompany.append(j.text)
            
    for k in exp:
        lexp.append(k.text)
                    
    for l in salary:
        lsalary.append(l.text)
            
    for m in location:
        llocation.append(m.text)
        
    for n in desc:
        ldesc.append(n.text)
        
    for o in req:
        lreq.append(o.text)
            
#         lcompany.append(j.text)
#         lexp.append(k.text)
#         lsalary.append(l.text)
#         llocation.append(m.text)
#         ldesc.append(n.text)
#         lreq.append(o.text)
    print('ROLE : ' ,len(lrole))
    print('Company : ',len(lcompany))
    print('exp : ',len(lexp))
    print('Salary : ',len(lsalary))
    print('location : ',len(llocation))
    print('Desc : ',len(ldesc))
    print('Req : ',len(lreq))
    
#     print('ROLE : ' ,lrole)
#     print('Company : ',lcompany)
#     print('exp : ',lexp)
#     print('Salary : ',lsalary)
#     print('location : ',llocation)
#     print('Desc : ',ldesc)
#     print('Req : ',lreq)
```

```python
flag = 0
count = 0
url = 'https://www.naukri.com/data-analyst-data-scientist-jobs-146?k=data%20analyst%2C%20data%20scientist'
#url = 'https://www.naukri.com/data-analyst-data-scientist-jobs-2'
lrole = list()
lcompany = list()
lexp = list()
lsalary = list()
llocation = list()
ldesc = list()
lreq = list()
driver = webdriver.Chrome('chromedriver.exe')
driver.get(url)
driver.maximize_window()
while(1):
    count = count + 1
    if flag == 0:
        findelement()
        time.sleep(4)
        try:
            print('count',count)
            driver.find_element_by_xpath('//a[@class="fright fs14 btn-secondary br2"]').click()
#             if count == 2:
#                 break
        except:
            flag = 1 
    else:
        break        
```
